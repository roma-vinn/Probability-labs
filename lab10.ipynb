```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
TRIALS = 10000
```

```python
def bernoulli_distr_LLN(p=0.5, trials=TRIALS):
    s = np.random.random(size=(trials-1))
    s = np.where(s>0.5, 1, 0)
    s = s.cumsum()
    y = s/np.arange(trials)[1:]
    return y
```

```python
Y = bernoulli_distr_LLN(0.3, 1000)
plt.plot(Y)
plt.ylim(0, 1)
plt.show()
```

```python
from scipy.stats import norm, uniform, expon
```

```python
SIZE = 1000
```

```python
x = norm.rvs(size=SIZE)
```

```python
y = np.maximum.accumulate(np.abs(x)) / np.sqrt(np.arange(1, SIZE + 1))
```

```python
plt.plot(y)
plt.show()
```

```python
def f(x):
    return x**2
```

```python
n = 10000
```

```python
# интеграл от 0 до 1 x^2
x = uniform.rvs(size=n)
x = f(x)
x = x.cumsum() / np.arange(1, n + 1)
plt.plot(x)
plt.show()
```

```python
def g(x):
    return 1/x**3
```

```python
# интеграл от 0 до 1 x^2
x = expon.rvs(size=n, loc=1)
x = g(x)/expon.pdf(x, loc=1)
x = x.cumsum() / np.arange(1, n + 1)
plt.plot(x)
plt.show()
```
